```python
import pandas as pd
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import MinMaxScaler
```

```python
# Read Data
data =  pd.read_csv("german_credit.csv", sep=",")
data
```

```python
y = data["Foreign Worker"]
x = data.drop(["Foreign Worker"], axis = 1)
```

```python
def SVM(x_data, y_data):
    # split the dataset into 0.6 trainig and 0.4 testing 
    x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size = 0.4, random_state = 0)
    # train the SVM classifier with linear kernel
    SVM_clf = LinearSVC()
    SVM_clf.fit(x_train,y_train)
    y_predict = SVM_clf.predict(x_test)
    accuracy = accuracy_score(y_test, y_predict)
    return accuracy
```

```python
# repeat training process 10 times and take average accuracy of 10 accuracies
def SVM_avg(x_data, y_data):
    accuracy_arr = []
    for i in range(10):
        accuracy_arr.append(SVM(x_data, y_data))
    accuracy_avg = np.average(accuracy_arr)
    return accuracy_avg
```

```python
# call fn and get result of average accuracy
print(" Average accuracy of data = ", SVM_avg(x, y))
```

```python
# preprocessing steps to standardize and normalize the dataset
x_scaled = x
# standardize features by subtracting the mean and scaling to unit variance
for i in x_scaled:
    Mean = np.mean(x_scaled[i])
    x_scaled[i] = x_scaled[i] - Mean
    
# normalization all dataset to range -1, 1
scaler = MinMaxScaler(feature_range=(-1, 1))
scaler.fit(x_scaled)
x_scaled = scaler.transform(x_scaled)
```

```python
# repeat the SVM model with new scaled data
print("Average accuracy of scaled data = ", SVM_avg(x_scaled, y))
```

The scaled data is better than the non scaled data. The accuracy of the scaled data is higher. Since the svm considers the changes of 1 to be constant with respect to multiple features. the change of one is significant. therefore the scaled data gives better accuracy
